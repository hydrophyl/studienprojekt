```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("feinstaub_14.2.csv", sep = ";")
```

```python
datum = df.timestamp[0].split('T')[0]
datentime = df.timestamp.str.split('T')
zeit = []
for dt in datentime:
    zeit.append(dt[1][:-3])

print(datum)
#Thêm cột zeit vào dataframe df
df['zeit'] = zeit
df_plot = df.loc[df.P1 < 2, ["zeit", "P1", "P2"]]
```

```python
x = df_plot.zeit
y = df_plot.P1
fig_1 = plt.figure(figsize=(25,3),dpi=200)
axes_1 = fig_1.add_axes([0,0,0.9,0.9])
axes_1.set_xlabel('Zeit')
axes_1.set_ylabel('PM2.5 (µg/m³)')
axes_1.set_title('< 2µg/m³ PM2.5')
axes_1.plot(x,y)
plt.bar(x,y)
```

```python
print("Summe P1: " , df.sum(axis=0)["P1"] , "µg/m³")
print("Summe P2: " , df.sum(axis=0)["P2"] , "µg/m³")
```
